```python
from bs4 import BeautifulSoup
import requests
import csv
import time
```

```python
# Base collection

base_url = '''https://www.sports-reference.com/cbb/play-index/pgl_finder.cgi?request=1&match=game&year_min=2015&
year_max=2019&school_id=&opp_id=&person_id=&is_range=N&game_type=A&game_month=&game_location=&game_result=&
game_num_min=&game_num_max=&is_starter=&pos_is_g=Y&pos_is_gf=Y&pos_is_f=Y&pos_is_fg=Y&pos_is_fc=Y&
pos_is_c=Y&pos_is_cf=Y&c1stat=&c1comp=&c1val=&c2stat=&c2comp=&c2val=&c3stat=&c3comp=&c3val=&c4stat=&
c4comp=&c4val=&is_dbl_dbl=&is_trp_dbl=&order_by=date_game&order_by_asc=&offset='''

game_stat_tags_leftclass = ['player', 'date_game', 'school_id', 'opp_id', 'game_result']

game_stat_tags_centerclass = ['pos']

game_stat_tags_rightclass = ['gs', 'mp', 'fg', 'fga', 'fg_pct', 'fg2', 'fg2a', 'fg2_pct', 'fg3', 'fg3a', 'fg3_pct', 
                             'ft', 'fta', 'ft_pct', 'game_score', 
                             'orb', 'drb', 'trb', 'ast', 'stl', 'blk', 'tov', 'pf', 'pts']

for offset in range(0,500,100): # Take this up to range(0, 575_000, 100) when ready
    url = base_url + str(offset)
    page = requests.get(url)
    soup = BeautifulSoup(page.text)
    
    my_dict_of_souplists = {}

    for stat in game_stat_tags_leftclass:
        my_dict_of_souplists[stat] = soup.find_all('td', {'class': 'left', 'data-stat': stat})

    for stat in game_stat_tags_centerclass:
        my_dict_of_souplists[stat] = soup.find_all('td', {'class': 'center', 'data-stat': stat})

    for stat in game_stat_tags_rightclass:
        my_dict_of_souplists[stat] = soup.find_all('td', {'class': 'right', 'data-stat': stat})
        
    print('Length of rows =', len(my_dict_of_souplists['player'][0]))
        
    for row_num in range(100):
        list_to_write = []
        for key in my_dict_of_souplists:
            list_to_write.append(my_dict_of_souplists[key][row_num].text)

        with open('./rolling_games.csv', 'a') as file:
            csv_output = csv.writer(file, delimiter=',')
            csv_output.writerow(list_to_write)
            
    time.sleep(1)
```
